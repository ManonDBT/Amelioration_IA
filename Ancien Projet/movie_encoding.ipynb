## importation des modules

```python
import pandas as pd
import numpy as np
```

## Importation du DF à encoder

```python
#lecture de la DF
df = pd.read_table('movie.csv', sep=',')
df
```

## Mettre tous les genres dans différentes colonnes 

```python
df['liste_genre']= df['genre'].str.split(',')
df
```

```python
def check_genre(nom_genre, x):
    if nom_genre in x :
        return 1
    else: 
        return 0 
check_genre("Drama","Adventude, Fantasy")
```

```python
df["Action"] = df["genre"].apply( lambda x: check_genre("Action", x))
df["Drama"] = df["genre"].apply( lambda x: check_genre("Drama", x))
df["Sci-Fi"] = df["genre"].apply( lambda x: check_genre("Sci-Fi", x))
df["Advendture"] = df["genre"].apply( lambda x: check_genre("Adventure", x))
df["Fantasy"] = df["genre"].apply( lambda x: check_genre("Fantasy", x))
df["Comedy"] = df["genre"].apply( lambda x: check_genre("Comedy", x))
df["History"] = df["genre"].apply( lambda x: check_genre("History", x))
df["Horror"] = df["genre"].apply( lambda x: check_genre("Horror", x))
df["Mystery"] = df["genre"].apply( lambda x: check_genre("Mystery", x))
df["Thriller"] = df["genre"].apply( lambda x: check_genre("Thriller", x))
df["Crime"] = df["genre"].apply( lambda x: check_genre("Crime", x))
df["Animation"] = df["genre"].apply( lambda x: check_genre("Animation", x))
df["Family"] = df["genre"].apply( lambda x: check_genre("Family", x))
df["Musical"] = df["genre"].apply( lambda x: check_genre("Musical", x))
df["Romance"] = df["genre"].apply( lambda x: check_genre("Romance", x))
df["Sport"] = df["genre"].apply( lambda x: check_genre("Sport", x))
df["Biography"] = df["genre"].apply( lambda x: check_genre("Biography", x))
df["Music"] = df["genre"].apply( lambda x: check_genre("Music", x))
df["Western"] = df["genre"].apply( lambda x: check_genre("Western", x))

df
```

## Séparer l'imdb en 2 parties. 

Cette séparation servira pour le machine learning afin d'obtenir que 2 catégories (les films à succès >7 et les autres <6)

```python
def check_imdb(i,x):
    if i <= x : 
        return 1
    else: 
        return 0 
check_imdb(7,8)
```

```python
df["imdb_encoding"] = df["imdb"].apply( lambda x: check_imdb(7, x))
df
```
